### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
# Your code here

def divide_numbers(x,y):
    try:
        result = x/y
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed."
    else:
        return result

# Test the function with different inputs
x = 10
y = 0

print(f"Result of {x} / {y}: {divide_numbers(x, y)}")
```


### Exercise 2: Raise ValueError for Invalid Input
Write a Python program that prompts the user to input an integer and raises a `ValueError` exception if the input is not a valid integer.



```python
# Your code here

def get_integer_input():
    try:
        user_input = input("Please enter an integer: ")
        number = int(user_input)
        print(f"You entered the integer: {number}")
    except ValueError:
        print("Error: That is not a valid integer.")

get_integer_input()
```



### Exercise 3: Handle FileNotFoundError
Write a Python program that opens a file and handles a `FileNotFoundError` exception if the file does not exist.



```python
# Your code here
def open_file(filename):
    try:
        with open(filename, 'r') as file:
            contents = file.read()
            print("File contents:\n")
            print(contents)
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")

filename = input("Enter the name of the file to open: ")
open_file(filename)
```



### Exercise 5: Handle PermissionError
Write a Python program that opens a file and handles a `PermissionError` exception if there is a permission issue.




```python
def open_file(filename):
    try:
        with open(filename, 'r') as file:
            contents = file.read()
            print("File contents:\n")
            print(contents)
    except PermissionError:
        print(f"Error: You don't have permission to read the file '{filename}'.")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")

filename = input("Enter the name of the file to open: ")
open_file(filename)
```



### Exercise 6: Handle IndexError in List Operations
Write a Python program that executes an operation on a list and handles an `IndexError` exception if the index is out of range.




```python
def access_list_element():
    my_list = [10, 20, 30, 40, 50]
    try:
        index = int(input(f"Enter an index (0 to {len(my_list)-1}): "))
        print(f"Element at index {index}: {my_list[index]}")
    except IndexError:
        print("Error: The index you entered is out of range.")
    except ValueError:
        print("Error: Please enter a valid integer.")

access_list_element()
```

Write a Python program that prompts the user to input a number and handles a `KeyboardInterrupt` exception if the user cancels the input.



```python
# Your code here
def safe_input():
    try:
        name = input("Please enter your name: ")
        print(f"Hello, {name}!")
    except KeyboardInterrupt:
        print("\nProgram was interrupted by the user (KeyboardInterrupt).")

safe_input()
```

Write a Python program that executes division and handles an `ArithmeticError` exception if there is an arithmetic error.



```python
def perform_calculation():
    try:
        a = float(input("Enter the first number: "))
        b = float(input("Enter the second number: "))
        result = a / b
        print(f"Result of {a} / {b} = {result}")
    except ArithmeticError as e:
        print(f"Arithmetic error occurred: {e}")
    except ValueError:
        print("Error: Please enter valid numbers.")

perform_calculation()
```



### Exercise 9: Handle UnicodeDecodeError
Write a Python program that opens a file and handles a `UnicodeDecodeError` exception if there is an encoding issue.



```python
def read_file_safely(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()
            print("File content:\n")
            print(content)
    except UnicodeDecodeError:
        print(f"Error: Cannot decode the file '{filename}' with UTF-8 encoding.")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")

filename = input("Enter the name of the file to read: ")
read_file_safely(filename)
```
