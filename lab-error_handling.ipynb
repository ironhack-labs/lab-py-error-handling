### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
# Your code here
def division(num1, num2):
    try:
        result = num1 / num2
        return result
    except ZeroDivisionError:
        print("Error: cannot divide by zero.")
        return None


print(division(10, 2))   # 5.0
print(division(10, 0))   # Error: cannot divide by zero.
```

```python
# Your code here
def verify_integer (x):
    try:
        if type(x) != int:
            raise ValueError
        print(f"Thank You! This is a valid integer number: {x}")  
    except ValueError:
        print(f"Error! This is not a valid integer number: {x}")

verify_integer(5)
verify_integer("number")
```

Write a Python program that prompts the user to input two numbers and raises a `TypeError` exception if the inputs are not numerical.



```python
# Your code here
def get_numerical_input():
    try:
        # Prompt user to input two values
        input1 = input("Enter the first number: ")
        input2 = input("Enter the second number: ")
        
        try:
            # Try to convert inputs to floats
            num1 = float(input1)
            num2 = float(input2)
        except ValueError:
            # Re-raise as a TypeError if conversion fails
            raise TypeError("Inputs must be numerical values.")
            
        print(f"Success! Both inputs are numerical: {num1} and {num2}")
        return num1, num2

    except TypeError as error:
        print(f"Caught an exception: {error}")

get_numerical_input()
```

```python
# Your code here
def append_to_number(number, value):
    try:
        number.append(value)
    except AttributeError:
        print(f"Error: '{type(number).__name__}' object has no attribute 'append'.")

append_to_number(3, 6)
```

Create a custom exception class and raise it in your code when certain conditions are met.




```python
# Your code here
# Step 1 — create a custom exception class
class NegativeNumberError(Exception):
    """Raised when a negative number is not allowed."""
    pass


# Step 2 — raise it when a condition is met
def square_root(number):
    try:
        if number < 0:
            raise NegativeNumberError(f"Cannot take square root of a negative number: {number}")
        return number ** 0.5
    except NegativeNumberError as e:
        print(f"Custom error: {e}")
        return None


print(square_root(25))    # 5.0  
print(square_root(-9))    # Custom error 
```

Write a program that repeatedly prompts the user for valid input until they provide it, using exception handling to manage invalid inputs.



```python
# Your code here
def get_valid_age() -> int:
    # Repeatedly prompts the user for a valid age until they provide one.
    
    while True:
        try:
            age = int(input("Enter your age: "))
            if age < 0:
                raise ValueError("Age cannot be negative.")
            if age > 120:
                raise ValueError("Age cannot be greater than 100.")
            # Returns: int: A valid age between 0 and 100.
            print(f"Valid age entered: {age}")
            return age
        except ValueError as e:
            print(f"Invalid input: {e}. Please try again.")

get_valid_age()
```

Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
# Your code here
def divide(max_attempts):
    #
    # Prompts the user for two numbers and divides them,
    # retrying up to max_attempts times on failure.
    #
    # Parameters:
    #    max_attempts (int): Maximum number of attempts. Defaults to 3.
    #

    attempts = 0

    while attempts < max_attempts:
        try:
            num1 = float(input("Enter the dividend: "))
            num2 = float(input("Enter the divisor: "))

            result = num1 / num2
            print(f"Success! {num1} / {num2} = {result}")
            return result

        except ArithmeticError as e:
            attempts += 1
            remaining = max_attempts - attempts
            print(f"Arithmetic error: {e}")
            if remaining > 0:
                print(f"Please try again. {remaining} attempt(s) remaining.\n")
            else:
                print("Maximum attempts reached. Exiting.")

        except ValueError:
            attempts += 1
            remaining = max_attempts - attempts
            print(f"Error: please enter valid numbers.")
            if remaining > 0:
                print(f"{remaining} attempt(s) remaining.\n")
            else:
                print("Maximum attempts reached. Exiting.")
    return None

divide(3)
```
