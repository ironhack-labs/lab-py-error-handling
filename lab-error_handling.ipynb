Write a Python program that opens a file and handles a `PermissionError` exception if there is a permission issue.




```python
# Your code here
def open_protected_file(file_path):
  try:
    with open(file_path, 'w') as f:
      f.write('Test data')

  except PermissionError:
    print("Error: You do not have sufficient permissions to access:")
```

```python
# Your code here
def safe_division():
    try:
        dividend = int(input("Enter the dividen: "))
        divisor = int(input("Enter the divisor: ")) 
        print(f"Result: {dividend/divisor}")
    # ArithmeticError 
    except ArithmeticError:
        print(f"Arithmetic error detected!!")

safe_division()
```

Write a Python program that opens a file and handles a `UnicodeDecodeError` exception if there is an encoding issue.



```python
# Your code here
def read_file_safely(file_path):
    try:

        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
            print("File read successfully")
            return content


    except UnicodeDecodeError:
        print("Encoding error (Encoding Error)")

# read_file_safely(file)
```
