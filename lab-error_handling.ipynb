### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
# Your code here
def div(num1,num2):
    try:
        result = num1 / num2
    except ZeroDivisionError:
    # This runs if num2 is zero
        print("Error: You cannot divide by zero!")
    else: 
        return result
```

```python
div(6,0)
```


### Exercise 2: Raise ValueError for Invalid Input
Write a Python program that prompts the user to input an integer and raises a `ValueError` exception if the input is not a valid integer.



```python
# Your code here
try:
    num1 = float(input("Enter a number: "))
    if num1 < 0:
        raise ValueError("Number must be positive")
    else:
        result = num1 ** 0.5   # example: take square root
    print("The result is:", result)
    
except ValueError as e:
    print("Error:", e)

# This saves the error message into a variable called e. "e as a little box that contains what went wrong."
```

Write a Python program that opens a file and handles a `PermissionError` exception if there is a permission issue.




```python
try:
    # Try to open the file in write mode ("w")
    with open("my_file.txt", "w") as file:
        file.write("Hello, world!")
    print("File written successfully.")

except PermissionError:
    # If Python does not have permission to write to this file
    print("Error: You do not have permission to write to this file.")
```

Write a Python program that prompts the user to input a number and handles a `KeyboardInterrupt` exception if the user cancels the input.



```python
try:
    # Keep asking the user for numbers
    while True:
        number = input("Type a number (or press Ctrl+C to stop): ")
        print("You typed:", number)

except KeyboardInterrupt:
    print("\nOkay! You stopped the program. Bye!")
```

Write a Python program that opens a file and handles a `UnicodeDecodeError` exception if there is an encoding issue.



```python
try:
    # Try to open a file and read it
    with open("my_file.txt", "r", encoding="utf-8") as file:
        content = file.read()
    print("File content:")
    print(content)

except UnicodeDecodeError:
    # This runs if Python cannot read the file because of weird characters
    print("Oops! The file has characters I can't understand.")
```
