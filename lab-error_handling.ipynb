Write a Python program that prompts the user to input two numbers and raises a `TypeError` exception if the inputs are not numerical.



```python
def get_numerical_input(prompt_message):
    user_input = input(prompt_message)
    try:
        # Attempt to convert to float to handle both integers and decimals
        number = float(user_input)
        return number
    except ValueError:
        # If conversion fails (i.e., not a valid numerical string),
        # raise a TypeError as specifically requested.
        raise TypeError(f"Input '{user_input}' is not numerical.")

try:
    print("Please enter two numbers.")
    number1 = get_numerical_input("Enter the first number: ")
    number2 = get_numerical_input("Enter the second number: ")
    print(f"You entered valid numerical inputs: {number1} and {number2}")
    # Example of an operation with numerical inputs
    # print(f"Their sum is: {number1 + number2}")
except TypeError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

Write a Python program that opens a file and handles a `UnicodeDecodeError` exception if there is an encoding issue.



```python
import os

file_name = 'encoded_file.txt'
incorrect_encoding_file = 'incorrect_encoded_file.txt'

# 1. Create a file with a specific encoding (e.g., latin-1)
try:
    with open(file_name, 'w', encoding='latin-1') as f:
        f.write("This is a test string with special character: ©")
    print(f"Successfully created '{file_name}' with latin-1 encoding.")
except Exception as e:
    print(f"Error creating file: {e}")

# 2. Attempt to read the file with an incorrect encoding (e.g., utf-8)
# This should trigger a UnicodeDecodeError
try:
    print(f"\nAttempting to read '{file_name}' with utf-8 encoding...")
    with open(file_name, 'r', encoding='utf-8') as f:
        content = f.read()
        print("File content:")
        print(content)
except UnicodeDecodeError as e:
    print(f"Error: UnicodeDecodeError occurred. There was an encoding issue while reading the file. Details: {e}")
except FileNotFoundError:
    print(f"Error: The file '{file_name}' was not found.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
finally:
    # Clean up the created file
    if os.path.exists(file_name):
        os.remove(file_name)
        print(f"Cleaned up: Removed '{file_name}'.")
```

Write a Python program that executes an operation on an object and handles an `AttributeError` exception if the attribute does not exist.



```python
class MyClass:
    def __init__(self, name):
        self.name = name

def demonstrate_attribute_error():
    my_object = MyClass("TestObject")

    try:
        print(f"Accessing existing attribute 'name': {my_object.name}")
        # Attempt to access a non-existent attribute
        print(f"Attempting to access non-existent attribute 'age': {my_object.age}")
    except AttributeError as e:
        print(f"Error: AttributeError occurred. The attribute does not exist. Details: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

demonstrate_attribute_error()
```

Create a custom exception class and raise it in your code when certain conditions are met.




```python
class InvalidAgeError(Exception):
    """Custom exception for invalid age values."""
    def __init__(self, age, message="Age must be between 0 and 120."):
        self.age = age
        self.message = message
        super().__init__(self.message)

def set_age(age):
    if not isinstance(age, (int, float)):
        raise TypeError("Age must be a number.")
    if not (0 <= age <= 120):
        raise InvalidAgeError(age, f"Invalid age: {age}. Age must be between 0 and 120.")
    print(f"Age set successfully to: {age}")

# Demonstrate handling the custom exception
print("--- Testing with valid age ---")
try:
    set_age(30)
except (InvalidAgeError, TypeError) as e:
    print(f"Caught error: {e}")

print("\n--- Testing with invalid age (too high) ---")
try:
    set_age(150)
except (InvalidAgeError, TypeError) as e:
    print(f"Caught error: {e}")

print("\n--- Testing with invalid age (negative) ---")
try:
    set_age(-5)
except (InvalidAgeError, TypeError) as e:
    print(f"Caught error: {e}")

print("\n--- Testing with invalid type ---")
try:
    set_age("twenty")
except (InvalidAgeError, TypeError) as e:
    print(f"Caught error: {e}")
```

Write a program that repeatedly prompts the user for valid input until they provide it, using exception handling to manage invalid inputs.



```python
def get_valid_integer_input(prompt):
    while True:
        try:
            user_input = input(prompt)
            number = int(user_input)
            print(f"Valid input received: {number}")
            break  # Exit loop if input is valid
        except ValueError:
            print("Invalid input. Please enter a whole number.")
        except KeyboardInterrupt:
            print("\nOperation cancelled by user.")
            return None # Or raise the exception again if preferred
        except Exception as e:
            print(f"An unexpected error occurred: {e}")
            return None # Or raise the exception again if preferred
    return number

# Example usage:
print("Please enter a valid integer.")
valid_number = get_valid_integer_input("Enter an integer: ")

if valid_number is not None:
    print(f"Program continues with valid number: {valid_number}")
```

Modify your error handling to log errors to a text file instead of printing them to the console.



```python
import logging

# Configure logging to a file
log_file_name = 'error_log.txt'
logging.basicConfig(
    filename=log_file_name,
    level=logging.ERROR, # Log only ERROR level and above
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def perform_division_with_logging(numerator_str, denominator_str):
    try:
        numerator = float(numerator_str)
        denominator = float(denominator_str)
        result = numerator / denominator
        print(f"Division successful: {result}")
        return result
    except ValueError as e:
        error_message = f"Invalid input: Could not convert '{numerator_str}' or '{denominator_str}' to a number. Details: {e}"
        logging.error(error_message)
        print(f"An error occurred. Check '{log_file_name}' for details.")
    except ZeroDivisionError as e:
        error_message = f"Arithmetic error: Attempted to divide {numerator_str} by zero. Details: {e}"
        logging.error(error_message)
        print(f"An error occurred. Check '{log_file_name}' for details.")
    except Exception as e:
        error_message = f"An unexpected error occurred during division. Details: {e}"
        logging.error(error_message)
        print(f"An error occurred. Check '{log_file_name}' for details.")
    return None

print("--- Testing valid division ---")
perform_division_with_logging("10", "2")

print("\n--- Testing division by zero ---")
perform_division_with_logging("10", "0")

print("\n--- Testing invalid input ---")
perform_division_with_logging("abc", "2")

print(f"\nAll errors have been logged to '{log_file_name}'.")

# Optionally, you can read and print the log file content here to verify
# try:
#     with open(log_file_name, 'r') as f:
#         print(f"\nContent of '{log_file_name}':\n---\n{f.read()}---")
# except FileNotFoundError:
#     print(f"Log file '{log_file_name}' not found yet.")
```

Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
def reliable_division():
    while True:
        try:
            num1_str = input("Enter the numerator (or 'q' to quit): ")
            if num1_str.lower() == 'q':
                print("Exiting retry logic.")
                break

            num2_str = input("Enter the denominator: ")

            num1 = float(num1_str)
            num2 = float(num2_str)

            result = num1 / num2
            print(f"Result of division: {result}")
            break  # Exit loop if division is successful

        except ValueError:
            print("Invalid input. Please enter valid numerical values. Let's try again.")
        except ZeroDivisionError:
            print("Error: Cannot divide by zero. Let's try again.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}. Let's try again.")

print("--- Demonstrating retry logic for division ---")
reliable_division()
```
