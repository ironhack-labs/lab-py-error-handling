Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
# Your code here
max_attempt = 3
current_attempt = 1

while (current_attempt <= max_attempt):
    print(f"attempt number {current_attempt}")
    inpt1 = input("Enter one number ")
    inpt2 = input("Enter another number ")
    try:
        inp_one = int(inpt1)
        inp_two = int(inpt2)
        print(inp_one + inp_two)
        max_attempt = 0
    except ValueError:
        current_attempt += 1
        print("value error, try again")
    
    
```
