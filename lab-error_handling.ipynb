### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
def secure_division(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: ZeroDivisionError"
        
secure_division(10, 0)
```


### Exercise 2: Raise ValueError for Invalid Input
Write a Python program that prompts the user to input an integer and raises a `ValueError` exception if the input is not a valid integer.



```python
def secure_int_prompt():
    try:
        return int(input("Please enter a integer number: "))
    except ValueError:
        return "Error: ValueError"
    
secure_int_prompt()
```



### Exercise 3: Handle FileNotFoundError
Write a Python program that opens a file and handles a `FileNotFoundError` exception if the file does not exist.



```python
def secure_open_file(location):
    try:
        with open(location, 'r', encoding="utf8") as f:
            return f.read()
    except FileNotFoundError:
        return "Error: FileNotFoundError"
    
secure_open_file('../not-exist.txt')
```



### Exercise 4: Raise TypeError for Non-Numerical Input
Write a Python program that prompts the user to input two numbers and raises a `TypeError` exception if the inputs are not numerical.



```python
def secure_numerical_prompt():
    try:
        a = input("Please enter first integer number: ")
        b = input("Please enter second integer number: ")
        return [i for i in range(a, b)]
    except TypeError:
        return "Error: TypeError"
    
secure_numerical_prompt()
```

Write a Python program that opens a file and handles a `PermissionError` exception if there is a permission issue.




```python
def secure_permission_file(location):
    try:
        with open(location, mode="w") as f:
            f.write("Write on read only")
            return True
    except PermissionError:
        return "Error: PermissionError"
    
secure_permission_file('data/permission-error.txt')
```



### Bonus Exercise 2: Create Custom Exception
Create a custom exception class and raise it in your code when certain conditions are met.




```python
# Inherit from exception class
class CatsException(Exception):
    def __init__(self, message="Cats not allowed"):
        self.message = message
        
def house_animal_welcome(animal):
    NOT_ALLOWED_ANIMALS = ['cat']
    if animal.lower() in NOT_ALLOWED_ANIMALS:
        raise CatsException()
    return f"Welcome {animal}!!"

def get_new_animal_on_house(animal):
    try:
        return house_animal_welcome(animal)
    except CatsException:
        return f"Never ever {animal}!!"
    
print(get_new_animal_on_house("Dog"))
print(get_new_animal_on_house("Cat"))
```

Write a program that repeatedly prompts the user for valid input until they provide it, using exception handling to manage invalid inputs.



```python
def validate_input():
    while True:
        try:
            num = int(input("Insert a number from 0 to 10: "))
            if num < 0 or num > 10:
                raise Exception
            else:
                return num
        except Exception:
            continue

validate_input()
```



### Bonus Exercise 4: Log Errors to File
Modify your error handling to log errors to a text file instead of printing them to the console.



```python
class NumberLimitException(Exception):
    MESSAGE = "Number range limit reached"
    def __init__(self, message=MESSAGE):
        super().__init__(self)
        self.message = message
        
    def __str__(self):
        return self.message

def validate_input():
    ERROR_LOG_LOCATION = './data/error-log.txt'
    while True:
        try:
            num = int(input("Insert a number from 0 to 10: "))
            if num < 0 or num > 10:
                raise NumberLimitException
            else:
                return num
        except Exception as e:
            with open(ERROR_LOG_LOCATION, mode="a") as f:
                f.write(str(e) + "\n")
            continue

validate_input()
```



### Bonus Exercise 5: Retry Logic on Exception
Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
def get_data_from_file():
    retry = True
    while retry:
        file_location = input("Type file location: ")
        try:
            with open(file_location, mode="r") as f:
                return f.read()
        except Exception:
            response = input("Do you want to try again (Y, N): ")
            if response.upper() == 'N':
                retry = False
            else:
                continue
            
get_data_from_file()
```
