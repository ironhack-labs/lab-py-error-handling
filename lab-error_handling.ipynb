### Bonus Exercise 1: Handle Multiple Exceptions
Write a Python program that demonstrates handling multiple exceptions in one block.




```python
# Your code here
def safe_divide(a, b):
    try:
        result = a / b
        print(f"{a} / {b} = {result}")
    except (ZeroDivisionError, TypeError, ValueError) as e:
        print(f"Error occurred: {type(e).__name__} - {e}")

# Test cases
safe_divide(10, 2)      # Works fine
safe_divide(10, 0)      # ZeroDivisionError
safe_divide("10", 2)    # TypeError
safe_divide(10, "2")    # TypeError
```

Create a custom exception class and raise it in your code when certain conditions are met.




```python
# Your code here
class NegativeNumberError(Exception):
    """Custom exception for negative numbers"""
    def __init__(self, value, message="Negative numbers are not allowed"):
        self.value = value
        self.message = message
        super().__init__(self.message)

def process_positive_number(num):
    if num < 0:
        raise NegativeNumberError(num)
    return num * 2

try:
    number = float(input("Enter a positive number: "))
    result = process_positive_number(number)
    print(f"Double your number: {result}")
except NegativeNumberError as e:
    print(f"Custom Error: {e.message} - got {e.value}")
except ValueError:
    print("Error: Please enter a valid number")
```

Write a program that repeatedly prompts the user for valid input until they provide it, using exception handling to manage invalid inputs.



```python
# Your code here
def get_valid_integer():
    while True:
        try:
            user_input = input("Please enter an integer: ")
            number = int(user_input)
            return number
        except ValueError:
            print("Invalid input! Please enter a valid integer.")
        except KeyboardInterrupt:
            print("\nProgram interrupted by user")
            return None

# Usage
valid_number = get_valid_integer()
if valid_number is not None:
    print(f"You entered: {valid_number}")
```

Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
# Your code here
def get_file_content_with_retry(max_attempts=3):
    attempts = 0
    
    while attempts < max_attempts:
        try:
            filename = input("Enter filename to open: ")
            with open(filename, 'r') as file:
                content = file.read()
                print("File content:")
                print(content)
                return content
        except FileNotFoundError:
            attempts += 1
            if attempts < max_attempts:
                print(f"File not found. {max_attempts - attempts} attempts remaining.")
            else:
                print("Maximum attempts reached. File not found.")
        except PermissionError:
            print("Permission denied. Please choose a different file.")
            break
        except Exception as e:
            print(f"Unexpected error: {e}")
            break
    
    return None

# Usage
get_file_content_with_retry()
```
