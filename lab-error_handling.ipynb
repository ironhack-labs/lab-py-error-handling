

### Bonus Exercise 2: Create Custom Exception
Create a custom exception class and raise it in your code when certain conditions are met.




```python
class CustomError(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

def check_number(num):
    if num < 0:
        raise CustomError("Negative numbers are not allowed.")
    elif num == 0:
        raise CustomError("Zero is not a valid number.")
    else:
        return f"The number is: {num}"

try:
    number = int(input("Enter a number: "))
    result = check_number(number)
    print(result)

except CustomError as e:
    print(f"Error: {e}")
except ValueError:
    print("Error: Please enter a valid integer.")
```

Write a program that repeatedly prompts the user for valid input until they provide it, using exception handling to manage invalid inputs.



```python
def get_valid_input():
    while True:
        try:
            num = int(input("Enter a valid integer: "))
            return num
        except ValueError:
            print("Error: Invalid input. Please enter a valid integer.")

# Calling the function to get valid input
valid_number = get_valid_input()
print(f"You entered a valid number: {valid_number}")
```

Modify your error handling to log errors to a text file instead of printing them to the console.



```python
import logging

# Set up the logging configuration to log errors to a file
logging.basicConfig(filename="error_log.txt", level=logging.ERROR, 
                    format="%(asctime)s - %(levelname)s - %(message)s")

def get_valid_input():
    while True:
        try:
            num = int(input("Enter a valid integer: "))
            return num
        except ValueError as e:
            # Log the error to a file
            logging.error(f"Invalid input: {e}")
            print("Error: Invalid input. Please enter a valid integer.")

# Calling the function to get valid input
valid_number = get_valid_input()
print(f"You entered a valid number: {valid_number}")
```



### Bonus Exercise 5: Retry Logic on Exception
Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
import logging

# Set up the logging configuration to log errors to a file
logging.basicConfig(filename="error_log.txt", level=logging.ERROR,
                    format="%(asctime)s - %(levelname)s - %(message)s")

def get_valid_input():
    attempts = 0
    max_attempts = 3  # Set the maximum number of attempts
    
    while attempts < max_attempts:
        try:
            num = int(input("Enter a valid integer: "))
            return num
        except ValueError as e:
            # Log the error to a file
            logging.error(f"Invalid input: {e}")
            attempts += 1
            print(f"Error: Invalid input. You have {max_attempts - attempts} attempts left.")
        
        # If the user exceeds max attempts, stop the loop
        if attempts >= max_attempts:
            print("Too many invalid attempts. Exiting program.")
            break

# Calling the function to get valid input
valid_number = get_valid_input()

if valid_number:
    print(f"You entered a valid number: {valid_number}")
```
