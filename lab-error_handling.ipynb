### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
def safe_divide(numerator, denominator):
    try:
        result = numerator / denominator
        return f"Result: {result}"
    except ZeroDivisionError:
        return "Error: You cannot divide a number by zero!"

# Test cases
print(safe_divide(10, 2))  # Valid division
print(safe_divide(10, 0))  # Triggers ZeroDivisionError
```

Write a Python program that opens a file and handles a `FileNotFoundError` exception if the file does not exist.



```python
def read_file(filename):
    try:
        # Attempt to open and read the file
        with open(filename, 'r') as file:
            content = file.read()
            return content
    except FileNotFoundError:
        # Handle the case where the file doesn't exist
        print(f"Error: The file '{filename}' was not found. Please check the filepath.")
        return "ERROR_NOT_FOUND"

# Test cases
print(read_file("imaginary_file.txt"))
```



### Exercise 4: Raise TypeError for Non-Numerical Input
Write a Python program that prompts the user to input two numbers and raises a `TypeError` exception if the inputs are not numerical.



```python
def get_numerical_inputs():
    # Prompt the user for input
    input1 = input("Enter the first number: ")
    input2 = input("Enter the second number: ")

    try:
        # Try to convert inputs to floats to verify they are numerical
        # If conversion fails, it naturally throws a ValueError, which we catch
        try:
            num1 = float(input1)
            num2 = float(input2)
        except ValueError:
            # We catch the ValueError and intentionally raise a TypeError as requested
            raise TypeError("Both inputs must be numerical values!")
        
        # If no error was raised, perform a simple calculation
        return f"Success! The sum of {num1} and {num2} is {num1 + num2}"

    except TypeError as error:
        # Catch and print our custom TypeError message
        return f"Caught an Exception: {error}"

# Run the program
print(get_numerical_inputs())
```



### Exercise 5: Handle PermissionError
Write a Python program that opens a file and handles a `PermissionError` exception if there is a permission issue.




```python
def modify_file(filename, new_data):
    current_content = read_file(filename)
    if current_content == "ERROR_NOT_FOUND":
        return f"Modify Error: Cannot modify '{filename}' because it does not exist."
    
    try:
        with open(filename, 'w') as file:
            file.write(new_data)
        return f"Success: File '{filename}' modified successfully!"
            
    except PermissionError:
        # This catches if you have permission to READ, but not to WRITE
        return f"Modify Error: You have read access, but write permission to '{filename}' was denied."


print(modify_file("READ_FILE_TEST", "modify data"))
print(modify_file("/etc/hosts", "new data")) 
```

Write a Python program that executes an operation on a list and handles an `IndexError` exception if the index is out of range.




```python
def get_list_element(items, index):
    try:
        # Attempt to access the item at the specified index
        requested_item = items[index]
        return f"Success! The item at index {index} is: {requested_item}"
        
    except IndexError:
        # Handle the error if the index is out of bounds
        return f"Error: Index {index} is out of range. Valid indexes for this list are 0 to {len(items) - 1}."

# Define a sample list
fruits = ["apple", "banana", "cherry", "date"]

# Test cases
print(get_list_element(fruits, 1))   # Valid index (prints "banana")
print(get_list_element(fruits, 10))  # Invalid index (Triggers IndexError)
```

Write a Python program that prompts the user to input a number and handles a KeyboardInterrupt exception if the user cancels the input.



```python
def input_with_cancel_handling():
    try:
        # Prompt the user for input
        print("Press Ctrl+C at any time to cancel.")
        user_number = input("Please enter a number: ")
        
        # If the user successfully inputs data without interrupting
        return f"Thank you! You entered the number: {user_number}"
        
    except KeyboardInterrupt:
        # Handle the manual cancellation gracefully
        print("\n") # Cleans up the terminal output line
        return "Operation cancelled by the user. Goodbye!"

# Run the program
print(input_with_cancel_handling())
```

Write a Python program that executes division and handles an `ArithmeticError` exception if there is an arithmetic error.



```python
# Your code here
def safe_math_operation(numerator, denominator):
    try:
        # Attempt the division
        result = numerator / denominator
        return f"Success! The result is: {result}"
        
    except ArithmeticError as error:
        # This catches ZeroDivisionError, OverflowError, etc.
        return f"Mathematical Error Detected: {error}"

# Test cases
print(safe_math_operation(10, 2))  # Valid math
print(safe_math_operation(10, 0))  # Triggers ZeroDivisionError (caught by ArithmeticError)
```

Write a Python program that opens a file and handles a `UnicodeDecodeError` exception if there is an encoding issue.



```python
def read_encoded_file(filename, standard_encoding='utf-8'):
    try:
        # Attempt to read the file with the specified encoding
        with open(filename, 'r', encoding=standard_encoding) as file:
            content = file.read()
            return f"Success! File content:\n{content}"
            
    except UnicodeDecodeError as error:
        # Handle the encoding mismatch gracefully
        return (
            f"Encoding Error: Failed to decode '{filename}' using {standard_encoding}.\n"
            f"Details: {error.reason} at position {error.start}-{error.end}.\n"
            f"Fix: Try opening the file with a different encoding (e.g., 'utf-16' or 'latin-1')."
        )


# Test cases
# This will crash into the UnicodeDecodeError because it's UTF-16 data being read as UTF-8
print(read_encoded_file("sample_utf16.txt", standard_encoding="utf-8"))

print("\n--- Fixing the issue ---")
# This will succeed because the encodings match
print(read_encoded_file("sample_utf16.txt", standard_encoding="utf-16"))
```

Write a Python program that executes an operation on an object and handles an `AttributeError` exception if the attribute does not exist.



```python
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        return f"{self.name} says Woof!"


def call_animal_action(animal_object, method_name):
    try:
        # getattr() looks up a method or attribute dynamically by its string name.
        # It's like typing: animal_object.method_name()
        action = getattr(animal_object, method_name)
        return action()
        
    except AttributeError:
        # Handle the case where the object doesn't have that method/attribute
        class_name = type(animal_object).__name__
        return f"Error: Objects of type '{class_name}' do not have a '{method_name}' method."


# Test Setup
my_dog = Dog("Buddy")

# Test cases
print(call_animal_action(my_dog, "bark"))  # Valid attribute/method
print(call_animal_action(my_dog, "fly"))   # Invalid attribute (Triggers AttributeError)
```
