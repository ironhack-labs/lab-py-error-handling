Write a Python program that opens a file and handles a `PermissionError` exception if there is a permission issue.




```python
def read_file(filename):
    
    try:
        with open(filename, "r") as file:
            content = file.read()
            print("File Content:\n", content)
    
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found. Please verify the filename and try again.")
    
    except PermissionError:
        print(f"Error: Permission denied. You do not have the required access to open '{filename}'.")
    
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    filename = input("Enter the filename: ")
    read_file(filename)
```



### Exercise 6: Handle IndexError in List Operations
Write a Python program that executes an operation on a list and handles an `IndexError` exception if the index is out of range.




```python
def access_list_element(lst, index):

    try:
        element = lst[index]
        print(f"Element at index {index}: {element}")
    
    except IndexError:
        print(f"Error: Index {index} is out of range. Please provide a valid index between 0 and {len(lst) - 1}.")
    
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    sample_list = [10, 20, 30, 40, 50]
    
    try:
        index = int(input("Enter the index to access: "))
        access_list_element(sample_list, index)
    except ValueError:
        print("Error: Please enter a valid integer index.")
```

Write a Python program that prompts the user to input a number and handles a `KeyboardInterrupt` exception if the user cancels the input.



```python
def get_user_input():
    """Prompts the user to enter a number and handles KeyboardInterrupt."""
    try:
        number = float(input("Enter a number: "))
        print(f"You entered: {number}")
    
    except KeyboardInterrupt:
        print("\nOperation canceled by the user. Exiting gracefully.")
    
    except ValueError:
        print("Error: Please enter a valid numerical value.")

if __name__ == "__main__":
    get_user_input()
```

Write a Python program that executes division and handles an `ArithmeticError` exception if there is an arithmetic error.



```python
def safe_division():

    try:
        numerator = float(input("Enter the numerator: "))
        denominator = float(input("Enter the denominator: "))
        
        result = numerator / denominator
        print(f"Result: {result}")

    except ArithmeticError:
        print("Error: An arithmetic error occurred. Division by zero is not allowed.")
    
    except ValueError:
        print("Error: Please enter valid numerical values.")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    safe_division()
```

Write a Python program that opens a file and handles a `UnicodeDecodeError` exception if there is an encoding issue.



```python
def read_file_safely(filename):
    try:
        with open(filename, "r", encoding="utf-8") as file:
            content = file.read()
            print("File Content:\n", content)

    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found. Please check the filename and try again.")

    except PermissionError:
        print(f"Error: Permission denied. You do not have the required access to open '{filename}'.")

    except UnicodeDecodeError:
        print(f"Error: Unable to decode '{filename}'. The file may be encoded in a different format. Try opening it with a different encoding.")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    filename = input("Enter the filename: ")
    read_file_safely(filename)
```

Write a Python program that executes an operation on an object and handles an `AttributeError` exception if the attribute does not exist.



```python
class SampleClass:
    def __init__(self):
        self.attribute = "I exist!"

def access_attribute(obj, attr_name):
    """Attempts to access an attribute of an object and handles AttributeError."""
    try:
        value = getattr(obj, attr_name)
        print(f"Value of '{attr_name}': {value}")

    except AttributeError:
        print(f"Error: The attribute '{attr_name}' does not exist on the object.")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    sample_object = SampleClass()
    attribute_name = input("Enter the attribute name to access: ")
    access_attribute(sample_object, attribute_name)
```

### Bonus Exercise 1: Handle Multiple Exceptions
Write a Python program that demonstrates handling multiple exceptions in one block.




```python
def process_user_input():
    """Prompts the user for a number and handles multiple exceptions."""
    try:
        user_input = input("Enter a number: ")
        number = float(user_input)  
        result = 10 / number  
        print(f"Result: {result}")

    except (ValueError, ZeroDivisionError) as e:
        if isinstance(e, ValueError):
            print("Error: Please enter a valid numerical value.")
        elif isinstance(e, ZeroDivisionError):
            print("Error: Division by zero is not allowed.")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    process_user_input()
```

Create a custom exception class and raise it in your code when certain conditions are met.




```python
class CustomError(Exception):
    """Custom exception class for specific error handling."""
    def __init__(self, message):
        super().__init__(message)

def check_positive_number(number):
    """Checks if the number is positive and raises CustomError if not."""
    if number < 0:
        raise CustomError("Error: The number must be positive.")

def process_user_input():
    """Prompts the user for a number and handles the custom exception."""
    try:
        user_input = input("Enter a positive number: ")
        number = float(user_input)
        check_positive_number(number)  # Check if the number is positive
        print(f"You entered a valid number: {number}")

    except CustomError as e:
        print(e)

    except ValueError:
        print("Error: Please enter a valid numerical value.")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    process_user_input()
```

Write a program that repeatedly prompts the user for valid input until they provide it, using exception handling to manage invalid inputs.



```python
def get_valid_number():
    """Repeatedly prompts the user for a valid positive number."""
    while True:
        try:
            user_input = input("Enter a positive number: ")
            number = float(user_input)  # Try to convert the input to a float
            
            if number <= 0:  # Check if the number is positive
                raise ValueError("The number must be positive.")
            
            return number  # Return the valid number if input is valid

        except ValueError as e:
            print(f"Error: {e}. Please try again.")

        except Exception as e:
            print(f"An unexpected error occurred: {e}. Please try again.")

if __name__ == "__main__":
    valid_number = get_valid_number()
    print(f"You entered a valid number: {valid_number}")
```

Modify your error handling to log errors to a text file instead of printing them to the console.



```python
import logging
logging.basicConfig(filename='error_log.txt', level=logging.ERROR, 
                    format='%(asctime)s - %(levelname)s - %(message)s')
def get_valid_number():
    """Repeatedly prompts the user for a valid positive number and logs errors."""
    while True:
        try:
            user_input = input("Enter a positive number: ")
            number = float(user_input) 
            
            if number <= 0:  # Check 
                raise ValueError("The number must be positive.")
            
            return number 

        except ValueError as e:
            logging.error(e)  # Log the error to the file
            print("Invalid input. Please try again.")

        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")  # Log unexpected errors
            print("An unexpected error occurred. Please try again.")

if __name__ == "__main__":
    valid_number = get_valid_number()
    print(f"You entered a valid number: {valid_number}")
```



### Bonus Exercise 5: Retry Logic on Exception
Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
def get_positive_number_with_retry(max_retries=3):
    retries = 0
    while retries < max_retries:
        try:
            user_input = input("Enter a positive number: ")
            number = float(user_input)  
            
            if number <= 0: 
                raise ValueError("The number must be positive.")
            
            return number 

        except ValueError as e:
            retries += 1
            print(f"Error: {e}. You have {max_retries - retries} attempts left. Please try again.")
        
        except Exception as e:
            retries += 1
            print(f"An unexpected error occurred: {e}. You have {max_retries - retries} attempts left. Please try again.")

    print("Max retries exceeded. Exiting.")
    return None  

if __name__ == "__main__":
    valid_number = get_positive_number_with_retry()
    if valid_number is not None:
        print(f"You entered a valid number: {valid_number}")
```
