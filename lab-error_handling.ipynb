### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
# Your code here
def divide_num(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed."
    except TypeError:
        return "Error: Invalid input type. Please provide numbers."
    else:
        return result  
divide_num(10, 0)
```


### Exercise 2: Raise ValueError for Invalid Input
Write a Python program that prompts the user to input an integer and raises a `ValueError` exception if the input is not a valid integer.



```python
# Your code here
def is_integer(x):
    try:
        a = int(x)
    except ValueError:
        raise ValueError(f"'{x}' is not a valid number.")
    else:
        return a
    
is_integer('a')
```

```python
# Your code here
def open_files(file_name):
    try:
        open(file_name, 'r')
    except FileNotFoundError:
        return f"Error: The file '{file_name}' was not found."
    else:
        return f"The file '{file_name}' was opened successfully."
open_files('non_existent_file.txt')
```

Write a Python program that prompts the user to input two numbers and raises a `TypeError` exception if the inputs are not numerical.



```python
# Your code here
def operations(x, y):
    try:
        addition = x + y
        subtraction = x - y
        multiplication = x * y
        division = x / y
    except TypeError:
        return "Error: Invalid input type. Please provide numbers."
    return addition, subtraction, multiplication, division
operations(10, 'a')
```

Write a Python program that opens a file and handles a `PermissionError` exception if there is a permission issue.




```python
# Your code here
def open_files(file_name):
    try:
        open(file_name, 'r')
    except FileNotFoundError:
        return f"Error: The file '{file_name}' was not found."
    except PermissionError:
        return f"Error: You do not have permission to open the file '{file_name}'."
    else:
        return f"The file '{file_name}' was opened successfully."
```

Write a Python program that prompts the user to input a number and handles a `KeyboardInterrupt` exception if the user cancels the input.



```python
# Your code here
import time

try:
    print("Program running... (press Ctrl + C to stop)")
    while True:
        print("Still working...")
        time.sleep(1)
except KeyboardInterrupt:
    print("\nProgram interrupted by user. Exiting cleanly...")
```
