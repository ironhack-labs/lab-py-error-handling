Create a custom exception class and raise it in your code when certain conditions are met.




```python
# Your code here
class NegativeNumberError(Exception):
    def __init__(self,number, message = "Negative numbers are not allowed"):
        self.number = number
        self.message = message
        super().__init__(self.message)

number = -2

try:
    if number < 0:
        print("Error!! negativ numbers are not accpeted, Please enter a valid positive integer")
    print("The Entered number is ", number)
except NegativeNumberError as e:
    print(f"The value {number} entered is not valid")
except ValueError:
    print("Please enter a valid number")
```

Write a program that repeatedly prompts the user for valid input until they provide it, using exception handling to manage invalid inputs.



```python
# Your code here
while True:
    try:
        number = 1
        if number <= 0:
            raise ValueError("The number must be positive.")
            print("Input with exception handling")
        break  # Exit the loop if input is valid
    except ValueError as e:
        print("Invalid input:", e)
    except KeyboardInterrupt:
        print("\nInput cancelled by user.")
        break

print("You entered:", number if 'number' in locals() else "No valid input provided.")


```

Modify your error handling to log errors to a text file instead of printing them to the console.



```python
# Your code here
def log_error(message, filename="error_log.txt"):
    """Append error messages to a log file."""
    with open(filename, "a") as f:
        f.write(message + "\n")

try:
    numerator = 2
    denominator = 0
    result = numerator / denominator
    print("Result:", result)

except ZeroDivisionError as e:
    log_error(f"ZeroDivisionError: {e}")
    print("An error occurred. Check 'error_log.txt' for details.")

except ValueError as e:
    log_error(f"ValueError: {e}")
    print("An error occurred. Check 'error_log.txt' for details.")

except Exception as e:
    log_error(f"Unexpected error: {e}")
    print("An unexpected error occurred. Check 'error_log.txt' for details.")


```



### Bonus Exercise 5: Retry Logic on Exception
Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
# Your code here
MAX_RETRIES = 3  # Maximum number of retries

for attempt in range(1, MAX_RETRIES + 1):
    try:
        numerator = 1
        denominator = 0
        result = numerator / denominator
        print("Result:", result)
        break  # Exit loop if successful
    except ValueError:
        print(f"Invalid input! Please enter numeric values. Attempt {attempt} of {MAX_RETRIES}.")
    except ZeroDivisionError:
        print(f"Cannot divide by zero. Attempt {attempt} of {MAX_RETRIES}.")
    except KeyboardInterrupt:
        print("\nInput cancelled by user.")
        break
else:
    print("Maximum attempts reached. Operation failed.")
```
