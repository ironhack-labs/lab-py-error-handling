Write a Python program that executes an operation on an object and handles an `AttributeError` exception if the attribute does not exist.



```python
# Your code here

class Person:
    def __init__(self, name):
        self.name = name

woman = Person("Alice")
print("Age:", woman.age)
```

```python
class Person:
    def __init__(self, name):
        self.name = name

# Create an instance of Person, which has a 'name' attribute but no 'age' attribute.
woman = Person("Alice")

try:
    # Try to access an attribute that doesn't exist (e.g., 'age').
    print("Age:", woman.age)
except AttributeError:
    # This block will execute if an AttributeError is raised.
    print("AttributeError caught")
```

```python
class Person:
    def __init__(self, name):
        self.name = name

# Create an instance of Person, which has a 'name' attribute but no 'age' attribute.
woman = Person("Alice")

try:
    # Try to access an attribute that doesn't exist (e.g., 'age').
    print("Age:", woman.age)
except AttributeError as error:
    # This block will execute if an AttributeError is raised.
    print("AttributeError caught:", error)
```
