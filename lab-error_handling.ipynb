Write a Python program that prompts the user to input an integer and raises a `ValueError` exception if the input is not a valid integer.



```python
# Your code here
def is_integer(s):
    try:
        int(s)
        return True
    except ValueError:
        return False

user_input = input("Please enter an integer: ")

if is_integer(user_input):
    int_value = int(user_input)  # Safe to convert to integer
    print(f"The integer value is {int_value}.")
else:
    print("The input is not a valid integer.")

```

Write a Python program that prompts the user to input two numbers and raises a `TypeError` exception if the inputs are not numerical.



```python
# Your code here
def is_numerical(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

def get_numerical_input(prompt):
    user_input = input(prompt)
    if not is_numerical(user_input):
        raise TypeError(f"The input '{user_input}' is not a valid number.")
    return float(user_input)

try:
    num1 = get_numerical_input("Please enter the first number: ")
    num2 = get_numerical_input("Please enter the second number: ")
    print(f"The first number is {num1} and the second number is {num2}.")
except TypeError as e:
    print(e)
```

Create a custom exception class and raise it in your code when certain conditions are met.




```python
# Your code here
class DivisionFromZero(Exception):
    """
    Exception raised for custom error scenarios.
    """
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

def divide(a, b):
    if a == 0:
        raise DivisionFromZero("Division from zero is not allowed")
    return a / b

divide(6, 3)
divide(0, 6)
```

Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python

# Your code here
import sys

def prompt_direction():
    MAX_RETRIES = 3
    current_tries = 0

    while current_tries <= MAX_RETRIES:
        inp = input("Enter a direction (W, A, S, D): ")
        if inp.lower() in ['w', 'a', 's', 'd']:
            print('You have moved...')
            return
        else:
            current_tries += 1
            print(f"Invalid input, {MAX_RETRIES - current_tries} attempts remaining.")
    
    print("aa! errors!")
    sys.exit()

prompt_direction()
```
