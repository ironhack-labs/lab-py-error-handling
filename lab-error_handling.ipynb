### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
def devision(a, b):
    if b == 0:
        raise ZeroDivisionError(" second number can not be 0")
    return a/b
```


### Exercise 2: Raise ValueError for Invalid Input
Write a Python program that prompts the user to input an integer and raises a `ValueError` exception if the input is not a valid integer.



```python
def integer():
    x = input("Please enter a number(integer)")
    try:
        return int(x)
    except:
        raise ValueError("Please only enter integers")
```

```python
integer()
```

Write a Python program that prompts the user to input two numbers and raises a `TypeError` exception if the inputs are not numerical.



```python
def two_numericals():
    x = input("Please enter a number(int or float)")
    y = input("Please enter a number(int or float)")
    try:
        float(x)
    except:
        raise TypeError("x not numerical")
    try:
        float(y)
        return float(x), float(y)
    except:
        raise TypeError("y not numerical")
```
