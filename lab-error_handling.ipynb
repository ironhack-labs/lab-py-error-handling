### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print(f"Error: cannot divide {a} by zero.")
        return None

# Demonstration
pairs = [(10, 2), (5, 0), (0, 3)]
for x, y in pairs:
    result = safe_divide(x, y)
    print(f"{x} / {y} = {result}")
```

Write a Python program that prompts the user to input an integer and raises a `ValueError` exception if the input is not a valid integer.



```python
def prompt_integer(prompt="Enter an integer: "):
    """
    Prompt the user for input and return it as an int.
    Raises ValueError if the input cannot be converted to an integer.
    """
    s = input(prompt)
    try:
        return int(s)
    except (ValueError, TypeError):
        raise ValueError(f"Invalid integer input: {s!r}")

# Demonstration: handle the raised ValueError here so the notebook doesn't stop.
try:
    user_int = prompt_integer("Please enter an integer: ")
    print("You entered:", user_int)
except ValueError as e:
    print("ValueError:", e)
```

Write a Python program that opens a file and handles a `FileNotFoundError` exception if the file does not exist.



```python
def read_file(path):
    """
    Attempt to open and read a text file.
    Handles FileNotFoundError and returns the file content or None.
    """
    try:
        with open(path, 'r', encoding='utf-8') as f:
            return f.read()
    except FileNotFoundError:
        print(f"FileNotFoundError: the file {path!r} does not exist.")
        return None

# Demonstration: create an example file and try reading two files (one missing)
with open("example.txt", "w", encoding="utf-8") as f:
    f.write("This is a sample file for Exercise 3.\n")

for path in ["nonexistent.txt", "example.txt"]:
    content = read_file(path)
    if content is not None:
        print(f"--- Contents of {path!r} ---\n{content}")
```

Write a Python program that prompts the user to input two numbers and raises a `TypeError` exception if the inputs are not numerical.



```python
def prompt_two_numbers(prompt1="Enter first number: ", prompt2="Enter second number: "):
    """
    Prompt the user for two values and return them as floats.
    Raises TypeError if either input cannot be converted to a number.
    """
    s1 = input(prompt1)
    s2 = input(prompt2)
    try:
        n1 = float(s1)
        n2 = float(s2)
    except (ValueError, TypeError):
        raise TypeError(f"Non-numerical input detected: {s1!r}, {s2!r}")
    return n1, n2

# Demonstration: catch the TypeError so the notebook doesn't stop if invalid input is given.
try:
    a, b = prompt_two_numbers()
    print("You entered:", a, b)
except TypeError as e:
    print("TypeError:", e)
```

Write a Python program that opens a file and handles a `PermissionError` exception if there is a permission issue.




```python
import os
import stat

def write_with_permission_handling(path, data="Sample text\n"):
    """
    Attempt to write to `path` and handle PermissionError if it occurs.
    """
    try:
        with open(path, 'w', encoding='utf-8') as f:
            f.write(data)
        print(f"Wrote to {path!r} successfully.")
    except PermissionError:
        print(f"PermissionError: cannot write to {path!r}. Check file permissions.")
    except Exception as e:
        print(f"Unexpected error writing to {path!r}: {e!r}")

# Demonstration: create a file, remove permissions, then try writing to trigger PermissionError on Unix-like systems.
demo_path = "protected_example.txt"
with open(demo_path, "w", encoding="utf-8") as f:
    f.write("initial content\n")

# Remove all permissions (Unix-like systems). This may not behave the same on Windows.
try:
    os.chmod(demo_path, 0)
except Exception:
    # If chmod isn't supported or fails, continue to attempt the write to demonstrate handling.
    pass

# This call should raise PermissionError on systems that honor the chmod above.
write_with_permission_handling(demo_path, "attempt to overwrite\n")

# Restore user read/write permission so the file can be cleaned up.
try:
    os.chmod(demo_path, stat.S_IRUSR | stat.S_IWUSR)
except Exception:
    pass

# Cleanup
try:
    os.remove(demo_path)
except Exception:
    pass
```

```python
# Exercise 6: Handle IndexError in List Operations
# This cell uses existing notebook variables `pairs`, `x`, and `y`.

def get_element(lst, idx, default=None):
    """Return lst[idx] or default if IndexError occurs."""
    try:
        return lst[idx]
    except IndexError:
        print(f"IndexError: index {idx} is out of range for list of length {len(lst)}.")
        return default

# Demonstrate safe access using existing variables
item_at_x = get_element(pairs, x, default=None)
print(f"Element at index {x} -> {item_at_x!r}")

item_at_y = get_element(pairs, y, default=None)
print(f"Element at index {y} -> {item_at_y!r}")

# Example of using the retrieved item only if it exists
if item_at_y is not None:
    print("Processing item_at_y:", item_at_y)
else:
    print("No item to process at index", y)
```

Write a Python program that opens a file and handles a `UnicodeDecodeError` exception if there is an encoding issue.



```python
# Your code here
def read_file_safely(filename):
    try:
        # Try opening with default encoding (usually 'utf-8')
        with open(filename, "r", encoding="utf-8") as file:
            content = file.read()
            print("File content read successfully:\n")
            print(content)
    except UnicodeDecodeError:
        print("UnicodeDecodeError: Could not decode file with UTF-8 encoding.")
        print("Retrying with 'latin-1' encoding...")
        try:
            with open(filename, "r", encoding="latin-1") as file:
                content = file.read()
                print("File content read successfully with latin-1:\n")
                print(content)
        except Exception as e:
            print(f"Failed to read file with fallback encoding. Error: {e}")
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


# Example usage
filename = "example.txt"  # Replace with your filename
read_file_safely(filename)
```

Write a Python program that executes an operation on an object and handles an `AttributeError` exception if the attribute does not exist.



```python
# Your code here
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def start(self):
        print(f"{self.brand} {self.model} is starting...")

# Create a Car object
my_car = Car("Toyota", "Corolla")

try:
    # Attempt to call a method that does not exist
    my_car.drive()
except AttributeError as e:
    print(f"AttributeError caught: {e}")
    print("The object does not have this attribute.")

# Continue with valid method
my_car.start()
```

Create a custom exception class and raise it in your code when certain conditions are met.




```python
# Your code here
# Define a custom exception
class NegativeNumberError(Exception):
    """Custom exception for handling negative numbers."""
    pass


def calculate_square_root(num):
    try:
        if num < 0:
            # Raise our custom exception
            raise NegativeNumberError("Cannot calculate square root of a negative number.")
        
        result = num ** 0.5
        print(f"The square root of {num} is {result}")
    
    except NegativeNumberError as e:
        print("Custom Exception Caught:", e)


# Test the function
calculate_square_root(25)   # Valid input
calculate_square_root(-9)   # Will raise the custom exception
```

Modify your error handling to log errors to a text file instead of printing them to the console.



```python
# Your code here
import logging

# Configure logging
logging.basicConfig(
    filename="errors.log",   # File to store logs
    level=logging.ERROR,     # Only log errors and above
    format="%(asctime)s - %(levelname)s - %(message)s"  # Timestamp, level, message
)

def divide(a, b):
    try:
        return a / b
    except Exception as e:
        logging.error("Error dividing %s by %s: %s", a, b, str(e))

# Example usage
print(divide(10, 2))   # Works fine
print(divide(5, 0))    # Logs error instead of printing
print(divide("a", 2))  # Logs
```

Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
# Your code here
import logging
import time

# Configure logging (from Exercise 4)
logging.basicConfig(
    filename="errors.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def divide_with_retry(a, b, retries=3, delay=2):
    """
    Try dividing a by b with retry logic.
    :param a: numerator
    :param b: denominator
    :param retries: number of attempts before giving up
    :param delay: seconds to wait before retry
    :return: result or None if failed
    """
    attempt = 0
    while attempt < retries:
        try:
            return a / b
        except Exception as e:
            attempt += 1
            logging.error("Attempt %s failed: %s", attempt, str(e))
            print(f"Error: {e}. Retrying ({attempt}/{retries})...")
            time.sleep(delay)  # wait before retry
    print("Operation failed after all retries.")
    return None

# Example usage
print(divide_with_retry(10, 2))   # ✅ Works
pr
```
