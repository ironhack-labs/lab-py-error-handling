### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
# Your code here

def divide_numbers(numerator, denominator):
    try:
        result = numerator / denominator
    except ZeroDivisionError:
        print("ZeroDivisionError: attempt to divide by zero.")
        return None

    return result

print("Divide result:", divide_numbers(5, 2))
print("Divide result:", divide_numbers(5, 0))
```

```python
# Your code here

def check_inputs():
    val1 = input("Please enter first number: ")
    val2 = input("Please enter second number: ")
    
    if (val1.isnumeric() == False or val2.isnumeric() == False):
        raise TypeError
    
    print(f"The {val1} and {val2} are numerical values.")

try:
    check_inputs()
except TypeError:
    print("TypeError: inputs not numerical values.")
```

Write a Python program that opens a file and handles a `PermissionError` exception if there is a permission issue.




```python
# Your code here

def open_file2(filename, permission_mode):
    try:
        fd = open(filename, permission_mode)
    except FileNotFoundError:
        print(f"FileNotFoundError: The file {filename} was not found.")
        return None
    except PermissionError:
        print(f"PermissionError: The file {filename} cannot be accessed for permissions {permission_mode}.")
        return None
    
    print(f"The file {filename} is opened.")
    return fd

def write_file(fd, text):
    try:
        fd.write(text)
    except PermissionError:
        print("PermissionError: writting not allowed.")
        return False
    
    return True

file3 = open_file2("test.txt", "w")
#file3 = open_file2("test2.txt", "w")
write_file(file3, "Cocojumbo") if file3 else None
file3.close() if file3 else None
```

Write a Python program that opens a file and handles a `UnicodeDecodeError` exception if there is an encoding issue.



```python
# Your code here
def open_file3(filename, permission_mode, encoding):
    try:
        fd = open(filename, permission_mode, encoding=encoding)
    except FileNotFoundError:
        print(f"FileNotFoundError: The file {filename} was not found.")
        return None
    except PermissionError:
        print(f"PermissionError: The file {filename} cannot be accessed for permissions {permission_mode}.")
        return None
    except UnicodeDecodeError:
        print(f"UnicodeDecodeError: The file {filename} cannot be opened as it is not a valid utf-8.")
        return None
    
    print(f"The file {filename} is opened.")
    return fd

def read_file(fd):
    try:
        line = fd.readline()
    except UnicodeDecodeError:
        print(f"UnicodeDecodeError: read failed as text is not a valid utf-8.")
        return None
    
    return line

file4 = open_file3("test3.txt", "r", "ascii")
print(read_file(file4) if file4 else None)
file4.close() if file4 else None
```

Write a program that repeatedly prompts the user for valid input until they provide it, using exception handling to manage invalid inputs.



```python
# Your code here

def prompting_user():
    while True:
        try:
            str = input("Enter a valid integer value:")
            print("You entered:", str)
            istr = int(str)

            return istr
        except KeyboardInterrupt:
            print("KeyboardInterrupt")
            continue
        except ValueError:
            print("ValueError")
            continue
        except TypeError:
            print("TypeError")
            continue


print("The entered value:", prompting_user())
```

Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
# Your code here

def prompting_user3():
    retry = True

    while retry == True:
        try:
            str = input("Enter a valid integer value:")
            istr = int(str)
            retry = False
        except KeyboardInterrupt as err:
            print(f"KeyboardInterrupt: {err}")
            continue
        except ValueError as err:
            print(f"ValueError: {err}")
            continue
        except TypeError as err:
            print(f"TypeError: {err}")
            continue

        return istr
        
print("The entered value:", prompting_user3())
```
