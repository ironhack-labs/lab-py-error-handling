### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
def error_handling(a, b):
    """
        This function takes two arguments, a and b, and attempts to divide a by b. 
        It includes error handling to catch potential exceptions that may arise during the division operation.
        Args:  a (int or float): The numerator.
        b (int or float): The denominator. 
        Returns: None: The function prints the result or error messages but does not return any value.
    """
    try:
        result = a / b
        print(f"The result of {a} divided by {b} is: {result}")
    except ZeroDivisionError:
        print("Error: You cannot divide by zero.")
    except TypeError:
        print("Error: Both inputs must be numbers.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


# Test cases
print(error_handling(10, 2))
print(error_handling(10, 0))
print(error_handling(10, "0"))
```

Write a Python program that opens a file and handles a `PermissionError` exception if there is a permission issue.




```python
# function to hndale permission error
def permission_error_handling(file_path):
    """
        This function attempts to read a file at the given file path and handles PermissionError if the file cannot be accessed due to insufficient permissions.
        Args: file_path (str): The path to the file to be read.
        Returns: None: The function does not return any value but prints the content of the file if it is successfully read or an error message if a PermissionError occurs.
    """
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            print(content)
    except PermissionError:
        print(f"PermissionError: You do not have permission to access the file at {file_path}.")
    except FileNotFoundError:
        print(f"FileNotFoundError: The file at {file_path} does not exist.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    
permission_error_handling("restricted_file.txt")

```

Write a Python program that executes an operation on a list and handles an `IndexError` exception if the index is out of range.




```python
# function to handle index error
def index_error_handling(lst, index):
    """
        This function takes a list and an index as input and attempts to access the element at the specified index. 
        It includes error handling to catch potential IndexError exceptions that may arise if the index is out of range.
        Args: lst (list): The list from which to access an element. index (int): The index of the element to access.
        Returns: None: The function prints the accessed element or an error message if an IndexError occurs.
    """
    try:
        element = lst[index]
        print(f"The element at index {index} is: {element}")
    except IndexError:
        print(f"IndexError: The index {index} is out of range for the list.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

index_error_handling([1, 2, 3, 4, 5], 2)
index_error_handling([1, 2, 3, 4, 5], 10)
```

Write a Python program that prompts the user to input a number and handles a `KeyboardInterrupt` exception if the user cancels the input.



```python
# function to handle keyboardInterrupt exception
import time

def keyboard_interrupt_handling():
    """
        This function demonstrates handling of KeyboardInterrupt exception.
        It includes a loop that can be interrupted by the user pressing Ctrl+C.
        Returns: None: The function does not return any value but prints a message if a KeyboardInterrupt is caught.
    """
    try:
        for i in range(10):
            print(f"Countdown: {10 - i}")
            time.sleep(1)
    except KeyboardInterrupt:
        print("\nKeyboardInterrupt: The operation was interrupted by the user.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

keyboard_interrupt_handling()
```

```python
# unicode error handling
def unicode_error_handling(text):
    """
        This function takes a string input and attempts to encode it in UTF-8 format. 
        It includes error handling to catch potential UnicodeError exceptions that may arise during the encoding process.
        Args: text (str): The input string to be encoded.
        Returns: None: The function prints the encoded string or an error message if a UnicodeError occurs.
    """
    try:
        encoded_text = text.encode('utf-8')
        print(f"The UTF-8 encoded version of the text is: {encoded_text}")
    except UnicodeError:
        print("UnicodeError: An error occurred during the encoding process.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

unicode_error_handling("Hello, World!")
unicode_error_handling("Hello, World! \udc80")
```

Write a Python program that executes an operation on an object and handles an `AttributeError` exception if the attribute does not exist.



```python
# handle attribute error
class SampleClass:
    def __init__(self, value):
        self.value = value

def attribute_error_handling(obj, attr_name):
    """
        This function takes an object and an attribute name as input and attempts to access the specified attribute. 
        It includes error handling to catch potential AttributeError exceptions that may arise if the attribute does not exist.
        Args: obj (object): The object from which to access an attribute. attr_name (str): The name of the attribute to access.
        Returns: None: The function prints the accessed attribute or an error message if an AttributeError occurs.
    """
    try:
        attr_value = getattr(obj, attr_name)
        print(f"The value of the attribute '{attr_name}' is: {attr_value}")
    except AttributeError:
        print(f"AttributeError: The attribute '{attr_name}' does not exist for the object.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

sample_obj = SampleClass(5)
attribute_error_handling(sample_obj, "value")
attribute_error_handling(sample_obj, "non_existent_attr")
```

Write a program that repeatedly prompts the user for valid input until they provide it, using exception handling to manage invalid inputs.



```python
# validate user input with exception handling
def validate_user_input(user_input):
    try:
        if not isinstance(user_input, str):
            raise ValueError("Input must be a string.")
        if len(user_input.strip()) == 0:
            raise ValueError("Input cannot be empty.")
        print(f"Valid input: {user_input}")
    except ValueError as ve:
        print(f"ValueError: {ve}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

try:
    while True:
        user_input = input("Please enter a string (or type 'exit' to quit): ")
        if user_input.lower() == 'exit':
            print("Goodbye!")
            break
        validate_user_input(user_input)
except KeyboardInterrupt:
    print("\nKeyboardInterrupt: Input cancelled by user.")
```

Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
# retry mechanism with exception handling
import time

def retry_with_backoff(func, max_retries=3, backoff_factor=1):
    for i in range(max_retries):
        try:
            return func()
        except Exception as e:
            logging.error(f"Attempt {i+1} failed: {e}")
            if i < max_retries - 1:
                time.sleep(backoff_factor * (2 ** i))
            else:
                raise Exception("All retry attempts failed.")

# Example usage
def example_function():
    # Simulate a function that may fail
    if time.time() % 2 < 1:  # Randomly fail half the time
        raise Exception("Simulated failure.")
    return "Success!"   

try:
    result = retry_with_backoff(example_function)
    print(result)
except Exception as e:
    print(f"Final error after retries: {e}")
```
