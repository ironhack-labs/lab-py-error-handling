### Bonus Exercise 1: Handle Multiple Exceptions
Write a Python program that demonstrates handling multiple exceptions in one block.




```python
# Your code here

def divide_user_input(a_str, b_str):
    """Convert two strings to numbers and divide them."""
    try:
        a = float(a_str)
        b = float(b_str)
        result = a / b
        return result
    except ValueError:
        print(f"   ❌  ValueError: '{a_str}' or '{b_str}' is not a number")
    except ZeroDivisionError:
        print(f"   ❌  ZeroDivisionError: cannot divide by zero")

# Test cases:
print(f"   ✅  10 / 2 = {divide_user_input('10', '2')}")
divide_user_input("hello", "2")
divide_user_input("10", "0")

# Option B — combined:
print("\n   Using combined tuple syntax:")
try:
    result = int("abc") / 0
except (ValueError, ZeroDivisionError) as e:
    print(f"   ⚠️ Error type is {type(e).__name__}: {e}")

```

Create a custom exception class and raise it in your code when certain conditions are met.




```python
# Your code here

# Step 1: Define the custom exception
class AgeError(Exception):
    """Raised when an age value is unrealistic."""
    def __init__(self, age, message="Age must be between 0 and 100"):
        self.age = age
        self.message = message
        super().__init__(self.message)

# Step 2: Use it in your code
def set_age(age):
    if age < 0 or age > 100:
        raise AgeError(age, f"Invalid age: {age}. Must be between 0 and 100.")
    print(f"   ✅  Age set to {age}")

# Test:
try:
    set_age(25)          # Valid
    set_age(200)         # Invalid → raises AgeError
except AgeError as e:
    print(f"   ❌  AgeError: {e}")
```

```python
# Your code here


import logging

# Configure logging to write to a file
log_filename = "error_log.txt"
logging.basicConfig(
    filename=log_filename,
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s",
    force=True     # Reset any existing config
)

# Generate some errors and log them:
try:
    result = 10 / 0
except ZeroDivisionError as e:
    logging.error(f"ZeroDivisionError: {e}")
    print(f"   📝  Logged ZeroDivisionError to '{log_filename}'")

try:
    number = int("not_a_number")
except ValueError as e:
    logging.error(f"ValueError: {e}")
    print(f"   📝  Logged ValueError to '{log_filename}'")

# Show the log file contents:
if os.path.exists(log_filename):
    with open(log_filename, "r") as f:
        print(f"\n   📄  Contents of {log_filename}:")
        for line in f:
            print(f"      {line.strip()}")
    os.remove(log_filename)   # Clean up
```
