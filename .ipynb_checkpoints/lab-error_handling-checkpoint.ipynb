### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
# Your code here
def division(a,b):
    '''
    Divides two numbers and handles zero division
    '''
    try:
        return a / b
    except ZeroDivisionError as e:
        raise ZeroDivisionError("b must be different than 0") from e


print(division(2,3))
print(division(2,0))
```


### Exercise 2: Raise ValueError for Invalid Input
Write a Python program that prompts the user to input an integer and raises a `ValueError` exception if the input is not a valid integer.



```python
# Your code here
def your_integer():
    '''
    takes an integer input and prints it or raises an error, prompting that the choice was wrong
    '''
    integer = input("Type your integer: ")

    try:
        number = float(integer)
    except ValueError:
        raise ValueError("That is not a number")

    if not number.is_integer():
        raise ValueError("That is not an integer")

    print(f"You chose {int(number)}")
your_integer()
```

Write a Python program that executes an operation on a list and handles an `IndexError` exception if the index is out of range.




```python
# Your code here
lst = [1,2,3,4,5]

def out_of_range_list(lista):
    try:
        for i in range(len(lista)+10):
            print(lista[100])
    except IndexError:
        raise

out_of_range_list(lst)
```

Write a program that repeatedly prompts the user for valid input until they provide it, using exception handling to manage invalid inputs.



```python
# Your code here

while True:
    try:
        x=int(input("Input 8: "))
        if x==8:
            break
        print("Try again!")
    except ValueError:
        print("That's not a number. Try again!")
```

Modify your error handling to log errors to a text file instead of printing them to the console.



```python
# Your code here
while True:
    try:
        x = int(input("Input 8: "))
        if x == 8:
            break
        with open("errors.txt", "a") as f:
            f.write(f"Wrong number entered: {x}\n")
    except ValueError:
        with open("errors.txt", "a") as f:
            f.write("Invalid input (not an integer)\n")
```



### Bonus Exercise 5: Retry Logic on Exception
Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
# Your code here

def retry(tries=2):
    while tries>0:
        try:
            x=int(input("Input 8: "))
            if x==8:
                break
            print("Try again!")
            tries -= 1
        except ValueError:
            print("That's not a number. Try again!")
            tries -= 1
```

```python
retry()
```
