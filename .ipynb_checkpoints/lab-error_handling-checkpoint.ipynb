### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
def divide(a, b):
    try:
        result = a / b
        return result 
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
        return None
```

```python
divide(5,0)
```


### Exercise 2: Raise ValueError for Invalid Input
Write a Python program that prompts the user to input an integer and raises a `ValueError` exception if the input is not a valid integer.



```python
products = ["t-shirt","mug","hat","book","keychain"]
def initialize_inventory(products):
    inventory= {}
    for product in products:
        try:
            quantity = int(input(f"Enter a quantity for the {product}:"))
            if quantity <0:
                raise ValueError
            inventory[product] = quantity 
        except ValueError:
            print("invalid quantity, product skipped")
        else:
            print("quantity updated")
        finally: print("initialization complete!")
    return inventory
```

```python
initialize_inventory(products)
```



### Exercise 3: Handle FileNotFoundError
Write a Python program that opens a file and handles a `FileNotFoundError` exception if the file does not exist.



```python
import pandas as pd
```

```python
def data_loader(path):
    try:
        df=pd.read_csv(path)
    except FileNotFoundError:
        print("invalid path!")
    return df
```

```python
data_loader("askjdhaskd.csv")
```

Write a Python program that prompts the user to input two numbers and raises a `TypeError` exception if the inputs are not numerical.



```python
def divide(a, b):
    try:
        result = a / b
    except TypeError:
        print("Error: Both arguments must be numbers.")
        return None
```

```python
divide(5, "one")
```

Write a Python program that executes an operation on a list and handles an `IndexError` exception if the index is out of range.




```python
numbers = [1, 2, 3]
try:
    print(numbers[2])     
    print(numbers[6])     
except IndexError:
    print("Error: Index is out of range!")
```

Write a Python program that prompts the user to input a number and handles a `KeyboardInterrupt` exception if the user cancels the input.



```python
try:
    name = input("Enter your name (or Ctrl+C to cancel): ")
    print(f"Hello, {name}!")
except KeyboardInterrupt:
    print("\nInput cancelled by user")
```

```python
def divide_numbers(a, b):
    try:
        result = a / b
        return result
    except ArithmeticError:
        return "Cannot perform this math operation"
```
