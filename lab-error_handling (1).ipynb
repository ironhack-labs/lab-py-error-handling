### Exercise 1: Handle ZeroDivisionError
Write a Python program to handle a `ZeroDivisionError` exception when dividing a number by zero.


```python
def divide_numbers():
    try:
        # Ask the user for inputs
        x = float(input("Enter the first number: "))
        y = float(input("Enter the second number: "))
        
        # Perform the division
        z = x / y
        print(f"The result is: {z}")
    
    except ZeroDivisionError:
        # Catch the exception if denominator is zero
        print("Cannot divide by zero! Please enter a valid denominator.")
    
    except ValueError:
        # Handle invalid inputs
        print("Invalid input! Please enter numeric values.")
        
divide_numbers()
```


### Exercise 2: Raise ValueError for Invalid Input
Write a Python program that prompts the user to input an integer and raises a `ValueError` exception if the input is not a valid integer.



```python
def get_integer_in_range(min_value, max_value):
    try:
        # Ask the user to input an integer
        user_input = input(f"Please enter an integer between {min_value} and {max_value}: ")
        
        # Try converting the input to an integer
        value = int(user_input)
        
        # Check if the value is within the specified range
        if value < min_value or value > max_value:
            raise ValueError(f"Input must be between {min_value} and {max_value}. You entered {value}.")
        
        print(f"You entered a valid integer: {value}")
    
    except ValueError as e:
        # If the input is not a valid integer or is out of the specified range
        print(e)
        
get_integer_in_range(1, 100)
```

Write a Python program that opens a file and handles a `FileNotFoundError` exception if the file does not exist.



```python
def open_file(filename):
    try:
        # Try to open the file
        with open(filename, 'r') as file:
            content = file.read()
            print("File content:")
            print(content)
    except FileNotFoundError:
        # Handle the case when the file is not found
        print(f"Error: The file '{filename}' was not found. Please check the file name and try again.")

# Call the function with a file name
open_file('Hello, My name is Deema!.txt')
```

Write a Python program that prompts the user to input two numbers and raises a `TypeError` exception if the inputs are not numerical.



```python
def get_numbers():
    try:
        # Ask the user to input two numbers
        x = input("Enter the first number: ")
        y = input("Enter the second number: ")
        
        # Check if the inputs are numerical
        if not (x.isdigit() and y.isdigit()):
            raise TypeError("Both inputs must be numerical values.")
        
        # Convert inputs to integers
        x = int(x)
        y = int(y)
        
        # Display the sum
        print(f"The sum of {x} and {y} is {x + y}")
    
    except TypeError as e:
        # Handle TypeError exception
        print(f"Error: {e}")

get_numbers()
```

Write a Python program that opens a file and handles a `PermissionError` exception if there is a permission issue.




```python
def open_file_with_permission_check(filename):
    try:
        # Try to open the file for writing
        with open(filename, 'w') as file:
            file.write("Hello, this is a test.")
            print("File written successfully.")
    
    except PermissionError:
        # Handle the case where there is a permission error
        print(f"Error: You do not have permission to write to the file '{filename}'. Please check your file permissions.")

open_file_with_permission_check('Hello, My name is Deema!.txt')
```

Write a Python program that executes an operation on a list and handles an `IndexError` exception if the index is out of range.




```python
def access_list_element(my_list, index):
    try:
        # Attempt to access an element at the given index
        print(f"The element at index {index} is: {my_list[index]}")
    
    except IndexError:
        # Handle the case when the index is out of range
        print(f"Error: Index {index} is out of range for the list. Please provide a valid index.")

my_list = [10, 23, 60, 89, 100]

access_list_element(my_list, 10)
```

Write a Python program that prompts the user to input a number and handles a `KeyboardInterrupt` exception if the user cancels the input.



```python
def get_user_input():
    try:
        # Asking the user for input
        x = int(input("Please enter a number: "))
        print(f"You entered: {x}")
    
    except KeyboardInterrupt:
        # Handle the case when the user cancels the input (Ctrl+C)
        print("\nInput was interrupted. The program will now exit.")

get_user_input()
```

Write a Python program that executes division and handles an `ArithmeticError` exception if there is an arithmetic error.



```python
def perform_operation():
    try:
        # Asking the user to input two numbers
        numerator = float(input("Enter the numerator: "))
        denominator = float(input("Enter the denominator: "))
        
        # Perform some mathematical operation
        result = (numerator ** 2) / (denominator - 2)
        print(f"The result of the operation is: {result}")
    
    except ArithmeticError:
        # Handle arithmetic error
        print("Error: An arithmetic error occurred. Make sure the denominator is not 2 to avoid division by zero.")
        
perform_operation()
```

Write a Python program that opens a file and handles a `UnicodeDecodeError` exception if there is an encoding issue.



```python
def open_file(filename):
    try:
        # Try to open the file with the default encoding
        with open(filename, 'r') as file:
            content = file.read()
            print(content)
    except UnicodeDecodeError:
        # Handle the case when there is a UnicodeDecodeError
        print(f"Error: The file '{filename}' contains characters that can't be decoded.")

open_file('Hello, My name is Deema!.txt')

```

Create a custom exception class and raise it in your code when certain conditions are met.




```python
class NegativeNumberError(Exception):
    def __init__(self, message="Negative numbers are not allowed"):
        self.message = message
        super().__init__(self.message)

# Function that raises custom exception if a negative number is input
def check_positive_number(number):
    if number < 0:
        raise NegativeNumberError("You entered a negative number!")
    else:
        print(f"Positive number: {number}")

try:
    check_positive_number(-5)
except NegativeNumberError as e:
    print(f"Error: {e}")
```

Write a program that repeatedly prompts the user for valid input until they provide it, using exception handling to manage invalid inputs.



```python
# Function to validate user input
def get_valid_input():
    while True:
        try:
            user_input = input("Please enter a valid integer: ")
            # Try to convert the input to an integer
            user_input = int(user_input)
            return user_input
        except ValueError:
            # Handle invalid input (non-integer)
            print("Invalid input! Please enter an integer.")
        
valid_number = get_valid_input()
print(f"Valid input received: {valid_number}")

```

Modify your error handling to log errors to a text file instead of printing them to the console.



```python
import logging

# Set up logging configuration to log errors to a file
logging.basicConfig(filename='error_log.txt', level=logging.ERROR)

def divide_numbers(a, b):
    try:
        result = a / b
        print(f"Result: {result}")
    except ZeroDivisionError as e:
        # Log the error to the file
        logging.error(f"Error: Tried to divide by zero. {e}")
        print("Error: Cannot divide by zero.")
    except Exception as e:
        # Log any other errors
        logging.error(f"Unexpected error: {e}")
        print("An unexpected error occurred.")

divide_numbers(5, 0)
divide_numbers(1007, 7)

```

Implement retry logic for operations that could fail, allowing users to try again after encountering an error.



```python
def get_integer():
    while True:
        try:
            number = int(input("Please enter an integer: "))
            print(f"You entered: {number}")
            break  # Exit the loop if input is valid
        except ValueError:
            print("That's not a valid integer. Please try again.")

get_integer()
```
